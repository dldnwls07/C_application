## 1. 소수 판별 알고리즘 (Prime Number Check)
* **관련 소스코드**: [prime.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/math/prime.c)

### 💡 핵심 개념
* **소수(Prime Number)**: 1보다 큰 자연수 중 1과 자기 자신만을 약수로 가지는 수.
* **시간 복잡도 최적화 ($\sqrt{N}$ 검사)**:
  * 어떤 수 $N$이 소수인지 판별하기 위해 2부터 $N-1$까지 전부 나누어 보는 방식은 **$O(N)$**의 시간이 걸립니다.
  * 그러나 $N$의 약수들은 대칭을 이루므로, **$2$부터 $\sqrt{N}$까지만 검사**하면 소수 여부를 완벽히 판별할 수 있습니다. 시간 복잡도가 **$O(\sqrt{N})$**으로 대폭 단축됩니다.
  * C언어에서는 실수 연산 오차가 발생할 수 있는 `sqrt()` 함수 대신, 정수 곱인 **`i * i <= num`** 조건을 주로 사용합니다.

### ⚠️ 암기 포인트
1. `num <= 1`인 경우는 소수가 아니므로 예외 처리를 먼저 수행합니다.
2. 루프 조건식은 `for (int i = 2; i * i <= num; i++)`로 설정합니다.
3. 나누어 떨어지는 순간 `is_prime = 0;`으로 설정하고 `break;`로 즉시 루프를 탈출합니다.


````c
#include <stdio.h>

int main() {
    int num;
    printf("숫자를 입력하세요: ");
    if (scanf("%d", &num) != 1) return 1;

    // 1. 예외 처리: 1 이하의 수는 소수가 아님
    if (num <= 1) {
        printf("소수가 아닙니다.\n");
        return 0;
    }

    int is_prime = 1; // 소수라고 가정
    
    // 2. 핵심 최적화 루프: 제곱근(sqrt)까지만 검사
    for (int i = 2; i * i <= num; i++) {
        if (num % i == 0) {
            is_prime = 0; // 약수를 찾았으므로 소수가 아님
            break;        // 더 이상 검사할 필요 없이 탈출
        }
    }

    // 3. 결과 출력
    if (is_prime) {
        printf("소수입니다.\n");
    } else {
        printf("소수가 아닙니다.\n");
    }

    return 0;
}
````

## 2. 거품 정렬 알고리즘 (Bubble Sort - Descending)
* **관련 소스코드**: [sort.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/sort/sort.c)

### 💡 핵심 개념
* **거품 정렬 (Bubble Sort)**: 인접한 두 개의 원소를 비교하여 크기 순서에 맞지 않으면 서로 자리를 바꾸는(Swap) 정렬 알고리즘.
* **루프 범위 제어**:
  * 원소 개수가 $N$개일 때, 1회전이 끝나면 가장 큰(또는 가장 작은) 원소가 맨 뒤로 고정됩니다.
  * 바깥 루프 `i`는 $0$부터 $N-2$까지 ($N-1$회전 진행).
  * 안쪽 루프 `j`는 $0$부터 $N-2-i$까지 돌며 비교합니다 (이미 확정된 뒷부분은 정렬 범위에서 제외).
* **내림차순(Descending) 조건**: `arr[j] < arr[j+1]` 일 때 Swap (앞 원소가 뒤 원소보다 작으면 자리를 교환하여 큰 값을 앞으로 보냄).

### ⚠️ 암기 포인트
1. 두 변수의 값을 서로 바꾸는 **Swap 공식**:
   ```c
   int temp = arr[j];
   arr[j] = arr[j + 1];
   arr[j + 1] = temp;
   ```
2. 정렬 안쪽 루프 끝 조건: **`N - 1 - i`** (암기 필수!)


````c
#include <stdio.h>

// 거품 정렬을 수행하는 함수 정의
void Sort() {
    int arr[7]; // 7개의 정수를 저장할 배열 선언
    int n = 7;  // 배열의 크기 (원소 개수)
    
    // ==========================================
    // 1. 사용자로부터 원소 입력 받기
    // ==========================================
    for (int i = 0; i < n; i++) {
        // 사용자에게 몇 번째 원소인지 안내 (i는 0부터 시작하므로 +1)
        printf("%d번째 원소값을 입력하세요 : ", i + 1);
        // 입력받은 정수를 배열의 i번째 칸에 저장
        scanf("%d", &arr[i]);
    }

    // ==========================================
    // 2. 거품 정렬 시작 (내림차순)
    // ==========================================
    // 바깥쪽 루프: 총 (원소개수 - 1)번만큼 크게 회전합니다.
    // 한 바퀴 돌 때마다 맨 뒤에 '최종 정렬된 고정석'이 하나씩 생깁니다.
    for (int i = 0; i < n - 1; i++) {
        
        // 안쪽 루프: 실제로 이웃한 원소끼리 크기를 비교하고 자리를 바꾸는 구간입니다.
        // - i 를 하는 이유: 이미 정렬이 끝난 맨 뒤의 고정석들은 비교할 필요가 없기 때문입니다.
        // - 1 을 하는 이유: j가 마지막 원소일 때 j+1(다음 원소)을 참조하면 배열 범위를 벗어나기 때문입니다.
        for (int j = 0; j < n - 1 - i; j++) {
            
            // 내림차순 정렬 조건: 현재 원소(앞)가 다음 원소(뒤)보다 '작으면' 자리를 바꿉니다.
            // (큰 수가 앞으로 오고, 작은 수는 자꾸 뒤로 밀려나게 됨)
            if (arr[j] < arr[j + 1]) {
                // [Swap 과정] 두 변수의 값을 안전하게 교환하기 위해 임시 변수(temp)를 사용합니다.
                int temp = arr[j];       // 1. 앞의 값을 temp에 백업
                arr[j] = arr[j + 1];     // 2. 뒤의 값을 앞의 칸에 덮어쓰기
                arr[j + 1] = temp;       // 3. 백업해둔 앞의 값을 뒤의 칸에 넣기
            }
        }
    }

    // ==========================================
    // 3. 정렬된 최종 결과 출력
    // ==========================================
    printf("\n내림차순 정렬 결과: ");
    // 배열의 0번째부터 6번째까지 차례대로 출력
    for (int i = 0; i < n; i++) {
        printf("%d ", arr[i]);
    }
    printf("\n"); // 줄바꿈으로 깔끔하게 마무리
}

// 프로그램의 시작점인 main 함수
int main() {
    Sort();  // 위에서 만든 정렬 함수를 호출하여 실행
    return 0; // 프로그램을 정상적으로 종료
}
````

## 3. 문자열 입력 및 분류 알고리즘 (String Handling & Counting)
* **관련 소스코드**: [strip_count.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/string/strip_count.c)

### 💡 핵심 개념
* **공백 포함 문자열 받기**: `scanf("%s", ...)`는 띄어쓰기(공백)나 탭을 만나면 입력을 종료합니다. 따라서 공백을 포함한 한 문장 전체를 입력받기 위해 **`fgets(str, sizeof(str), stdin)`**을 사용합니다.
* **줄바꿈 문자(`\n`) 처리**: `fgets()`는 사용자가 누른 엔터키(`\n`)를 버퍼에서 읽어와 문자열 끝에 포함시킵니다. 따라서 문자 개수를 세거나 비교할 때 `\n` 문자는 카운트에서 제외하는 예외 처리가 필요합니다.
* **문자 범위 체크**: 아스키 코드(ASCII) 값 비교 방식을 이용합니다:
  * **숫자 (0~9)**: `str[i] >= '0' && str[i] <= '9'`
  * **공백 (Space)**: `str[i] == ' '`

### ⚠️ 암기 포인트
1. `fgets` 사용 패턴: `fgets(배열명, 크기, stdin)`
2. `strlen(str)`로 전체 문자열의 길이를 구한 뒤 루프 실행.
3. 개행 문자 무시: `if (str[i] == '\n') continue;`


````c
#include <stdio.h>
#include <string.h>

int main() {
    char str[256];
    int digits = 0;
    int spaces = 0;
    int others = 0;

    printf("문자열 입력하기: ");
    // 1. 공백 포함 문자열 입력을 위해 fgets 사용
    fgets(str, sizeof(str), stdin);

    int len = strlen(str);
    for (int i = 0; i < len; i++) {
        // 2. fgets로 들어온 개행 문자(\n)는 분류에서 제외
        if (str[i] == '\n') {
            continue;
        }

        // 3. 아스키 범위 비교를 통한 분류
        if (str[i] >= '0' && str[i] <= '9') {
            digits++;
        } else if (str[i] == ' ') {
            spaces++;
        } else {
            others++;
        }
    }

    printf("숫자 개수 : %d\n", digits);
    printf("공백 개수 : %d\n", spaces);
    printf("그외 문자 개수 : %d\n", others);

    return 0;
}
````

## 4. 구조체와 값에 의한 전달 (Struct & Call-by-Value)
* **관련 소스코드**: [student.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/struct/structure/student.c)

### 💡 핵심 개념
* **구조체 (Struct)**: 다양한 종류(타입)의 변수들을 묶어 개발자가 새로 정의하는 사용자 지정 데이터 타입.
* **값에 의한 호출 (Call-by-Value)**:
  * C언어에서는 함수에 인자를 전달할 때 기본적으로 **값이 복사**되어 넘어갑니다.
  * 구조체를 매개변수로 전달하면 구조체 전체 필드가 고스란히 복사되므로, 함수 내부에서 멤버 변수를 변경하더라도 **호출부(`main`)의 원본 변수에는 아무런 지장을 주지 못합니다**.
  * 값을 함수 내부에서 수정하고 원본에 대입하려면, 반드시 함수가 **구조체 인스턴스를 직접 반환(return)**하고 이를 호출부에서 다시 받아 대입(`s = Get(s)`)해야 합니다.

### ⚠️ 암기 포인트
1. 구조체 타입 명시는 C언어에서 `struct 구조체명`으로 써야 합니다 (예: `struct Student s;`).
2. 값에 의한 전달을 사용할 때 함수의 반환 타입은 해당 구조체이어야 합니다: `struct Student Get(struct Student s)`
3. 호출 및 덮어쓰기 형태: `s = Get(s);`


````c
#include <stdio.h>

// 1. 구조체 정의
struct Student {
    int student_id;
    int korean;
    int english;
    int math;
    int total;
    double avg;
};

// 2. 구조체를 통째로 받아 복사본을 수정하고 다시 반환하는 함수 (Call-by-Value)
struct Student Get(struct Student s) {
    s.total = s.korean + s.english + s.math;
    s.avg = s.total / 3.0;
    return s; // 수정된 복사본을 반환
}

int main() {
    struct Student s;

    // 입력 받기
    printf("학번 : ");
    scanf("%d", &s.student_id);
    printf("국어 : ");
    scanf("%d", &s.korean);
    printf("영어 : ");
    scanf("%d", &s.english);
    printf("수학 : ");
    scanf("%d", &s.math);

    // 3. 값의 변경사항을 덮어씌우기 위해 대입 연산자 사용
    s = Get(s);

    // 결과 출력
    printf("\n학번 : %d\n", s.student_id);
    printf("국어 : %d\n", s.korean);
    printf("영어 : %d\n", s.english);
    printf("수학 : %d\n", s.math);
    printf("총점 : %d\n", s.total);
    printf("평균 : %.2f\n", s.avg);

    return 0;
}
````

## 5. 배수 필터링 및 실수 형변환 (Multiple Filtering & Casting)
* **관련 소스코드**: [mul3_mul7.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/basic_loop/mul3_mul7.c)

### 💡 핵심 개념
* **공배수 필터링**: 두 가지의 조건(예: 3의 배수이면서 7의 배수)을 모두 만족시키는 숫자는 나머지 연산자 `%`와 논리 AND 연산자 `&&`를 엮어 사용합니다: `(i % 3 == 0 && i % 7 == 0)`
* **실수 나눗셈 형변환 (Type Casting)**:
  * C언어에서 `int` 타입 변수들끼리 나눗셈을 하면 결과도 소수점 아래가 절삭된 `int`가 됩니다.
  * 예: `count / num1` (둘 다 정수이면 결과는 `0` 또는 `1` 등의 정수가 됨).
  * 백분율 등의 정확한 실수 값을 얻기 위해서는 계산 식 내부에서 최소 한 개의 변수를 `(double)` 또는 `(float)`로 캐스팅해주어야 실수 나눗셈이 작동합니다: `((double)count / total) * 100`

### ⚠️ 암기 포인트
1. 공배수 조건: `i % A == 0 && i % B == 0`
2. 나눗셈 결과의 실수를 얻으려면 캐스팅을 앞에 꼭 명시: `(double)count / total`


````c
#include <stdio.h>

int main() {
    int num1 = 0;
    int count = 0;
    
    printf("숫자 입력 : ");
    scanf("%d", &num1);

    if (num1 <= 0) {
        printf("0보다 큰 자연수를 입력하세요\n");
        return 1;
    }

    // 1. 공배수 판별 루프
    for (int i = 1; i <= num1; i++) {
        if (i % 3 == 0 && i % 7 == 0) {
            count++;
            printf("%d\n", i);
        }
    }

    // 2. 핵심: (double) 형변환으로 실수 나눗셈 진행
    double ratio = ((double)count / (num1 + 1) * 100);
    printf("%d개 비율은 %.2f%%\n", count, ratio);
    
    return 0;
}
````

## 6. 루프 보폭 조정을 통한 조건 분기 제거 (Odd & Even Sum)
* **관련 소스코드**: [odd_add_even.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/basic_loop/odd_add_even.c)

### 💡 핵심 개념
* **조건문 없는 홀짝 계산**:
  * 1부터 100까지 숫자를 하나씩 돌면서 `if (i % 2 == 1)` 인지 판별하는 방식도 틀린 것은 아닙니다.
  * 하지만 루프의 **증감부(`i += 2`)**를 조절하여 홀수만 혹은 짝수만 방문하게 하면, 조건문 분기를 거치지 않으므로 실행 속도가 빠르고 코드가 직관적이게 됩니다.
  * **홀수만 순회**: `for (int i = 1; i <= 100; i += 2)`
  * **짝수만 순회**: `for (int i = 2; i <= 100; i += 2)`

### ⚠️ 암기 포인트
1. 홀수 루프의 시작점은 `i = 1`, 짝수 루프의 시작점은 `i = 2`입니다.
2. 증감 연산자는 `i++`가 아닌 **`i += 2`**를 사용합니다.


````c
#include <stdio.h>

void process() {
    int sum_odd = 0;
    int sum_even = 0;

    // 1. 홀수 합 계산 (보폭 2)
    for (int i = 1; i <= 100; i += 2) {
        sum_odd += i;
    }

    // 2. 짝수 합 계산 (보폭 2)
    for (int i = 2; i <= 100; i += 2) {
        sum_even += i;
    }

    printf("홀수의 합 %d\n", sum_odd);
    printf("짝수의 합 %d\n", sum_even);
}

int main() {
    process();
    return 0;
}
````
