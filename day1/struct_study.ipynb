## 1. 구조체 (Structure)

### 💡 개념
* **구조체**는 서로 다른 자료형의 변수들을 묶어 하나의 의미 있는 자료 단위로 만든 사용자 정의 자료형입니다.
* 배열이 **동일한 자료형**의 데이터 모임이라면, 구조체는 **다른 자료형**의 데이터도 하나로 묶을 수 있습니다.

### 🔍 주요 연산자 및 접근 방법
1. **멤버 참조 연산자 온점(`.`)**: 구조체 일반 변수에서 내부 멤버 변수에 접근할 때 사용합니다. (`struct_var.member`)
2. **멤버 참조 연산자 화살표(`->`)**: 구조체 포인터 변수에서 가리키는 구조체의 멤버 변수에 접근할 때 사용합니다. (`struct_ptr->member`는 `(*struct_ptr).member`와 동일)
3. **값에 의한 전달 (Call-by-Value)**: 구조체를 함수의 인자로 넘기면 전체 필드가 복사되어 들어갑니다. 함수 내부의 변경이 호출부 원본에 반영되지 않으므로, 반환값으로 구조체를 리턴하거나 구조체 포인터를 넘겨야 합니다.

### [예제 12.5] 구조체의 중첩 사용 (Nested Struct)

### [실습문제 12.2] 중첩 구조체를 이용한 직선 길이 구하기
* 파일 경로: [px12_2.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/struct/structure/px12_2.c)
* 설명: 두 점 `p1`, `p2`를 갖는 `Line` 구조체를 활용하여 피타고라스 정리와 `sqrt()` 함수로 직선 길이를 계산합니다.
\n

```c
#include <stdio.h>
#include <math.h> // sqrt() 함수 사용

struct Point {
    int x;
    int y;
};

struct Line {
    struct Point p1;
    struct Point p2;
};

int main() {
    struct Line ln;
    float lx, ly;
    double length;

    ln.p1.x = 4;
    ln.p1.y = 6;
    ln.p2.x = 14;
    ln.p2.y = 16;

    // 두 점 사이의 가로, 세로 거리 계산
    lx = (float)(ln.p2.x - ln.p1.x);
    ly = (float)(ln.p2.y - ln.p1.y);
    
    // 피타고라스 정리를 이용한 두 점 사이의 직선 길이 계산
    length = sqrt(lx * lx + ly * ly);

    printf("p1의 좌표 : (%d, %d)\n
", ln.p1.x, ln.p1.y);
    printf("p2의 좌표 : (%d, %d)\n
", ln.p2.x, ln.p2.y);
    printf("ln의 길이 : %lf \n
", length);

    return 0;
}
\n
```

### [예제 12.6] 구조체 포인터와 화살표 연산자
* 파일 경로: [ex12_6.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/struct/structure/ex12_6.c)
* 설명: 구조체 주소를 가리키는 구조체 포인터 변수에서 멤버에 접근하기 위한 `(*ptr).x` 형식과 화살표 연산자 `ptr->x`를 학습합니다.
\n

```c
#include <stdio.h>

struct Point {
    int x;
    int y;
    int z;
};

typedef struct Point point;

int main() {
    point A = {3, 4, 5};
    point B = {2, 6, 9};
    point *ptr1, *ptr2;

    ptr1 = &A;
    ptr2 = &B;

    // 1. 일반 구조체 변수로 멤버 접근 (.)
    printf("A점의 좌표 : (%d, %d, %d) \n
", A.x, A.y, A.z);
    
    // 2. 구조체 포인터 변수의 역참조 후 멤버 접근 (*ptr1).member
    printf("A점의 좌표 : (%d, %d, %d) \n
", (*ptr1).x, (*ptr1).y, (*ptr1).z);
    
    // 3. 구조체 포인터 변수로 직접 멤버 접근 (-> 화살표 연산자)
    printf("B점의 좌표 : (%d, %d, %d) \n
", ptr2->x, ptr2->y, ptr2->z);

    return 0;
}
\n
```

### [실습문제 12.3] 구조체 포인터를 활용한 직선 길이 계산
* 파일 경로: [px12_3.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/struct/structure/px12_3.c)
* 설명: 앞서 구한 직선 길이 계산법에 구조체 포인터와 `->` 연산자를 결합해 구현합니다.
\n

```c
#include <stdio.h>
#include <math.h>

struct Point {
    int x;
    int y;
};

struct Line {
    struct Point p1;
    struct Point p2;
};

int main() {
    struct Line ln;
    struct Line *pln = &ln; // 구조체 포인터 선언 및 원본 주소 대입
    float lx, ly;
    double length;

    // 구조체 포인터를 사용하여 값 대입
    pln->p1.x = 4;
    pln->p1.y = 6;
    pln->p2.x = 14;
    pln->p2.y = 16;

    // 구조체 포인터(->)를 활용해 원소 간의 연산 수행
    lx = (float)(pln->p2.x - pln->p1.x);
    ly = (float)(pln->p2.y - pln->p1.y);
    length = sqrt(lx * lx + ly * ly);

    printf("p1의 좌표 : (%d, %d)\n
", pln->p1.x, pln->p1.y);
    printf("p2의 좌표 : (%d, %d)\n
", pln->p2.x, pln->p2.y);
    printf("ln의 길이 : %lf \n
", length);

    return 0;
}
\n
```

### [예제 12.7] 구조체의 함수 인수 전달 및 반환
* 파일 경로: [ex12_7.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/struct/structure/ex12_7.c)
* 설명: 키보드로부터 입력을 받아 구조체를 통째로 반환하고, 구조체 복사본을 함수 인수로 받아 출력하는 패턴을 공부합니다.
\n

```c
#include <stdio.h>

typedef struct Student {
    int id;
    char name[10];
    int age;
} student;

// 함수 선언
void stuPrn(student temp);
student stuInput();

int main() {
    student stu1;
    
    // 함수 반환값(구조체)을 변수에 대입
    stu1 = stuInput();
    
    printf("\n
 학번 이름 나이 \n
");
    printf("============================\n
");
    
    // 구조체를 인수로 함수에 전달 (값에 의한 전달)
    stuPrn(stu1);
    printf("\n
");

    return 0;
}

// 구조체를 출력하는 함수 (값에 의한 전달로 복사본을 받음)
void stuPrn(student temp) {
    printf("%d %s %d \n
", temp.id, temp.name, temp.age);
}

// 키보드 입력을 받아 구조체 변수를 통째로 반환하는 함수
student stuInput() {
    student temp;
    printf("학번을 입력하세요: ");
    scanf("%d", &temp.id);
    
    printf("이름을 입력하세요: ");
    scanf("%s", temp.name);
    
    printf("나이를 입력하세요: ");
    scanf("%d", &temp.age);
    
    return temp;
}
\n
```

## 2. 공용체 (Union)

### 💡 개념
* **공용체**는 구조체와 동일한 외형을 지니나, 내부의 모든 멤버 변수가 **동일한 메모리 공간을 공유**하는 점이 결정적으로 다릅니다.
* 따라서, 공용체는 가장 기억 공간이 큰 멤버의 메모리 크기만큼만 할당받으며, 특정 한 시점에는 **어느 한 멤버의 데이터만 보관**할 수 있습니다.
* 한 데이터를 여러 타입으로 다르게 뷰(view)해서 사용하고 싶거나, 메모리를 효율적으로 아껴야 하는 시스템 레벨 코딩에 유용합니다.

---

### [예제 12.8] 공용체의 공유 메모리 원리 확인
* 파일 경로: [ex12_8.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/struct/union/ex12_8.c)
* 설명: 공용체에 값을 대입할 때마다 이전에 대입되어 있었던 다른 타입의 멤버 변수 값이 완전히 덮어씌워져 오염되는 것을 프린트하여 확인합니다.
\n

```c
#include <stdio.h>

// 공용체 정의: 모든 멤버가 메모리 공간을 공유함
union value {
    char c;
    int i;
    double d;
};

typedef union value number;

int main() {
    number num;

    // 1. char 대입 시
    num.c = 61; // 61은 아스키코드로 '=' 문자
    printf("num.c = %d 를 보관하는 경우\n
", num.c);
    printf(" num.c = %d, num.i = %d, num.d = %e \n
", num.c, num.i, num.d);

    // 2. int 대입 시 (이전 char 데이터 오염)
    num.i = 541494;
    printf("num.i = %d 를 보관하는 경우\n
", num.i);
    printf(" num.c = %d, num.i = %d, num.d = %e \n
", num.c, num.i, num.d);

    // 3. double 대입 시 (이전 int 데이터 오염)
    num.d = 3.141592;
    printf("num.d = %lf 를 보관하는 경우\n
", num.d);
    printf(" num.c = %d, num.i = %d, num.d = %e \n
", num.c, num.i, num.d);

    return 0;
}
\n
```

### [예제 12.9] 구조체와 공용체의 메모리 크기 비교
* 파일 경로: [ex12_9.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/struct/union/ex12_9.c)
* 설명: `sizeof` 연산자를 활용하여 구조체(멤버 합산 크기)와 공용체(가장 큰 멤버의 크기)의 메모리 할당 차이를 비교해 봅니다.
\n

```c
#include <stdio.h>

struct Student {
    int id;          // 4 bytes
    char name[16];   // 16 bytes
    int age;         // 4 bytes
}; // 구조체의 메모리 크기는 대략 멤버들의 합 (24 bytes)

union Register {
    char ch;         // 1 byte
    short int sh;    // 2 bytes
    long int lo;     // 8 bytes (64비트 시스템 기준)
}; // 공용체의 메모리 크기는 가장 큰 멤버의 크기 (8 bytes)

int main() {
    struct Student stu1;
    union Register reg1;

    printf("stu1의 메모리 크기 : %d \n
", (int)sizeof(stu1));
    printf("reg1의 메모리 크기 : %d \n
", (int)sizeof(reg1));

    // 메모리 공유 상태 확인
    reg1.lo = 0x12345678;
    printf("reg1.ch = %8x (가장 하위 1바이트만 출력)\n
", reg1.ch);
    printf("reg1.sh = %8x (하위 2바이트 출력)\n
", reg1.sh);
    printf("reg1.lo = %8lx \n
", reg1.lo);

    return 0;
}
\n
```

## 3. 열거체 (Enum)

### 💡 개념
* **열거체(Enumeration)**는 변수가 가질 수 있는 값들을 기호 상수로 미리 나열해 두고, 그 범위 외의 값이 대입되지 않도록 막는 일종의 안전장치 역할을 하는 자료형입니다.
* 가독성을 극대화하기 위해 `const`를 대체하는 용도로 흔히 쓰입니다.

### 🔍 정수값 할당 규칙
1. 별도의 초깃값을 정해주지 않으면, 첫 번째 상수는 **0**으로 할당되고 이후 기호 상수들은 차례로 **1씩 증가**합니다.
2. 특정 상수에 직접 정수값을 할당하면, 그 이후 상수들은 **명시된 값에서부터 다시 1씩 증가**합니다.

```c
#include <stdio.h>

// 열거형 선언: 값을 별도로 지정하지 않으면 0부터 1씩 증가하는 정수가 할당됨
enum seasons { spring, summer, autumn, winter };
// spring=0, summer=1, autumn=2, winter=3

// 열거자 정수를 텍스트로 보여주기 위한 맵핑 배열
char *s_name[] = {"spring", "summer", "autumn", "winter"};

int main() {
    enum seasons s;

    s = spring;
    printf("s : %d --> season : %s \n
", s, s_name[s]);

    s = winter;
    printf("s : %d --> season : %s \n
", s, s_name[s]);

    return 0;
}
\n
```

### [예제 12.11] 열거형 기호 상수의 값 수동 배정 및 활용
* 파일 경로: [ex12_11.c](file:///Users/woojin/TukoreaUnivStudy/1-1(summer)/C/day1/struct/enum/ex12_11.c)
* 설명: 특정 원소의 값을 수동으로 지정(`blue = 16`)했을 때 후속 원소들의 값이 어떻게 변하는지 보고, 열거자끼리 덧셈을 수행해 봅니다.
\n

```c
#include <stdio.h>

// 열거자들의 기본값 지정 예시
enum colors { 
    black,      // 0
    brown,      // 1
    red,        // 2
    orange,     // 3
    yellow,     // 4
    green,      // 5
    blue = 16,  // 16으로 명시
    violet,     // 17 (이전 값인 16에서 1 증가)
    gray,       // 18
    white       // 19
};

typedef enum colors color_band;

int main() {
    color_band code;
    int value;

    code = orange;
    printf("code = orange --> %d \n
", code);

    code = gray;
    printf("code = gray --> %d \n
", code);

    // 열거형 값은 정수형이므로 덧셈 등 수식 계산이 가능함
    value = red + yellow; // 2 + 4 = 6
    printf("value = red + yellow --> %d \n
", value);

    return 0;
}
\n
```
